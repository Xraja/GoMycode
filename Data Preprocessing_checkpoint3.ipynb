```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv('C:/Users/H.A/Downloads/titanic-passengers.csv', sep=";")
df.head()
```

```python
print(df. isnull().sum())

```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Cabin'].fillna('C91',inplace=True)
df['Embarked'].fillna('C',inplace=True)
print('verification null value', df. isnull().sum())
```

```python
df.info()
```

```python
df['Survived']=LabelEncoder().fit_transform(df['Survived'])
df['Sex']=LabelEncoder().fit_transform(df['Sex'])
df['Ticket']=LabelEncoder().fit_transform(df['Ticket'])
df['Cabin']=LabelEncoder().fit_transform(df['Cabin'])
df['Embarked']=LabelEncoder().fit_transform(df['Embarked'])
df.head()
```

```python
print(df.info())
print('verification null value', df. isnull().sum())
```
